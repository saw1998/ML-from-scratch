## Question 1

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
X_train=np.array(df_train['Feature'])
y_train=np.array(df_train[" Label"])
X_test=np.array(df_test['Feature'])
y_test=np.array(df_test[' Label'])
```

```python
# print("Train Data")
plt.figure()
plt.scatter(X_train,y_train)
plt.title("Training Data")
plt.show()
# print("Test Data")
plt.figure()
plt.scatter(X_test,y_test)
plt.title("Test Data")
plt.show()
```

```python
alpha=0.05
m1=X_train.size
m2=X_test.size
# Errors=[]
hypothesis_curve_parameters=[]
```

```python
for n in range(1,10):
    thetas=0.5*np.ones(n+1)                                                      
    Total_J_theta=[]
    itrn=3000
    for itr in range(itrn):
      # hi = (h - y)*x
        hi=np.zeros(thetas.size)    
        J=0
        for data in range(m1):
            h=0
            for l in range(thetas.size):
              h+=thetas[l]*(X_train[data]**l)
            h-=y_train[data]  
          
            J+=(h)**2
            for k in range(thetas.size):
                hi[k]=hi[k] + h*(X_train[data]**k)  
    
            
        Total_J_theta.append(J/(2*m1))
        for i in range(thetas.size):
            thetas[i]=thetas[i]-(alpha/m1)*hi[i]  


    h_train=[] 
    for x in X_train:
        h=0
        for l in range(thetas.size):
            h+=thetas[l]*(x**l)
        h_train.append(h)
    h_test=[] 
    for x in X_test:
        h=0
        for l in range(thetas.size):
            h+=thetas[l]*(x**l)
        h_test.append(h)
    
    print("\n********************For Training data***************************")
    print('*********  For n = ',n,' *********')
    plt.figure()
    plt.title("Train data and hypothesis function for n = "+ str(n))
    plt.scatter(X_train,y_train,label='Training')
    plt.scatter(X_train,h_train,label='hypothesis')
    plt.legend(loc='upper right' )
    plt.show()
    plt.figure()
    plt.title("Cost function v/s iteration")
    plt.plot(range(itr+1),Total_J_theta,)
    plt.show()

    print("##################################################################")
    print('*********  For n = ',n,' *********')
    print("Parameters are :")
    for i in range(thetas.size):
        print('     theta',i,' = ',thetas[i])

    test_error=0;
    for i in range(X_test.size):
        t=0
        for j in range(thetas.size):
              t+=thetas[j]*(X_test[i]**j)
        t-=y_test[i]
        test_error+=t**2
        # test_error+=(thetas[0]+thetas[1]*X_test[i]+ thetas[2]*(X_test[i]**2) + thetas[3]*(X_test[i]**3) +thetas[4]*(X_test[i]**4) + thetas[5]*(X_test[i]**5) -y_test[i])**2           #do changes  
    print('\nOn Training Dataset')
    test_error=test_error/(2*m1)
    print("Test error="  ,test_error)
    # Errors.append(test_error)
    hypothesis_curve_parameters.append(thetas)
    print()
    plt.figure()
    plt.scatter(X_test,y_test,label='Test')
    plt.scatter(X_test,h_test,label='hypothesis')
    plt.title("Test data and hypothesis function for n = "+str(n))
    plt.show()
    
    
    print('##################################################################')
```

## Question 2

```python
for n in range(0,9):
    x=np.linspace(0.0,1.0,200)
    hypo_fun=[]
    for i in x:
        h=0
        for j in range(hypothesis_curve_parameters[n].size):
              h+=hypothesis_curve_parameters[n][j]*(i**j)
        hypo_fun.append(h)
    plt.figure()
    plt.plot(x,hypo_fun)
    plt.title("for n = "+str(n+1))
    plt.show()
```

```python
J_train=[]
J_test=[]
for n in range(0,9):
    
    J_tn=0
    for x in range(X_train.size):
        tn=0
        for i in range(hypothesis_curve_parameters[n].size):
            tn+=hypothesis_curve_parameters[n][i]*(X_train[x]**i)
        tn-=y_train[x]
        J_tn+=tn*tn/(2*m1)
    J_train.append(J_tn)

    J_tt=0
    for x in range(X_test.size):
        tt=0
        for i in range(hypothesis_curve_parameters[n].size):
            tt+=hypothesis_curve_parameters[n][i]*(X_test[x]**i)
        tt-=y_test[x]
        J_tt+=tt*tt/(2*m2)
    J_test.append(J_tt)

iteration=np.linspace(1,9,9) 
plt.figure()
plt.plot(iteration,J_train)
plt.title("On Train Dataset")
plt.show()

plt.figure()
plt.plot(iteration,J_test)
plt.title("On test Dataset")
plt.show()
```

since the test-error is minimum for n=9**

## Question  3

```python
lamda=np.array([0.25,0.5,0.75,1.0])
n_max_err=2
n_min_err=9
```

## 3a

```python
#For max error

J_train=[] 
J_test=[]
thetas=0.5*np.ones(n_max_err)                                                     
itrn=3000
for lam in lamda:
    for itr in range(itrn):
      # hi = (h - y)*x
        hi=np.zeros(thetas.size)    
        for data in range(m1):
            h=0
            for l in range(thetas.size):
              h+=thetas[l]*(X_train[data]**l)
            h-=y_train[data]  
          
            for k in range(thetas.size):
                hi[k]=hi[k] + h*(X_train[data]**k)  

      
        for i in range(thetas.size):
            if i==0:
                thetas[i]=thetas[i]-(alpha/m1)*hi[i]  
            else:
                thetas[i]=thetas[i]-(alpha/m1)*hi[i] - lam/(2*m1)                                                  # for first part



    thetas_sum=0
    for i in thetas:
        thetas_sum+=i

    h1_train=0
    for x in range(X_train.size):
        h=0
        for l in range(thetas.size):
            h+=thetas[l]*(X_train[x]**l)
        h1_train+=(h-y_train[l])**2
    J_train.append(( (h1_train)/(m1) ))

    h1_test= 0
    for x in range(X_test.size):
        h=0
        for l in range(thetas.size):
            h+=thetas[l]*(X_test[x]**l)
        h1_test+=(h-y_test[l])**2
    J_test.append(( (h1_test)/(m2) ))

print("********************   For maximum error (n=2)   ***********************")
plt.figure()
plt.plot(lamda,J_train)
plt.title("Regularization with absolute parameters on Training Data")
plt.xlabel("lamda")
plt.ylabel("Error")
plt.show()

plt.figure()
plt.plot(lamda,J_test)
plt.title("Regularization with absolute parameters on Testing Data")
plt.xlabel("lamda")
plt.ylabel("Error")
plt.show()


```
